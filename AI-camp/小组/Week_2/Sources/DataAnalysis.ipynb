```python
import numpy as np

mean = np.mean(family, axis=0)
std = np.std(family, axis=0)
family = (family - mean) / std
family
```

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(family, target, test_size=0.2)
# 训练集中样本的个数
n_train = X_train.shape[0]
# 训练集中样本特征值的个数
n_features = X_train.shape[1]
```

```python
import MyProgress as mpg

modolone = mpg.MyRegression(method='ols')
modolone.fit(X_train, y_train)
```

```python
y_predict_one = modolone.predict(X_test)
# 线性回归拟合图示例
plt.figure(figsize=(8, 4))
sns.scatterplot(x=y_test, y=y_predict_one,s=6,color='yellow')
y_predict_one = modolone.predict(X_train)
sns.scatterplot(x=y_train, y=y_predict_one, s=6,color='blue')
y_predict_one = modolone.predict(X_test)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--')
plt.xlabel('True Values')
plt.ylabel('Predicted Values')
plt.title('Linear Regression Fit Plot')
plt.show()
# print("参数向量 theta:", modolone.theta)
print('mean_squared_error',modolone.mean_squared_error(y_test,y_predict_one))
print('score',modolone.score(X_test, y_test))
print('mean_absolute_error',modolone.mean_absolute_error(y_test,y_predict_one))
```

## 以上是用<u><span style="color:'green';">最小二乘法</span></u>的算法实现的线性回归模型,<span style="color:'yellow';">黄色</span>的点代表测试集，而<span style="color:'blue';">蓝色</span>的点代表训练集，看得出来拟合的还行

```python
def model_2_test(i) -> None:
    modoltwo = mpg.MyRegression(method='gd',max_iter=i)
    modoltwo.fit(X_train, y_train)

    y_predict_two = modoltwo.predict(X_test)
    # 线性回归拟合图示例
    plt.figure(figsize=(6, 2))
    sns.scatterplot(x=y_test, y=y_predict_two,s=3)
    plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--')
    plt.xlabel('True Values')
    plt.ylabel('Predicted Values')
    plt.title('Linear Regression Fit Plot')
    plt.show()
    # print("参数向量 theta:", modoltwo.theta)
    print("迭代",i,'次')
    print('mean_squared_error',modoltwo.mean_squared_error(y_test,y_predict_two))
    print('score',modoltwo.score(X_test, y_test))
    print('mean_absolute_error',modoltwo.mean_absolute_error(y_test,y_predict_two))
```

```python
model_2_test(10)
model_2_test(100)
model_2_test(500)
model_2_test(1000)
```

## 以上是用<u><span style="color:'green';">梯度下降法</span></u>训练得到的线性回归模型

#### 由以上两种算法可看出，对同一个测试集来说，<u><span style="color:'yellow';">最小二乘法和梯度下降法拟合的程度都差不多(迭代多次)</span></u>，但是梯度下降法如果迭代次数比较少(小于<u><span style="color:'green';">100</span></u>则会不准确)，但是由于学习度大小不变的原因，<u><span style="color:'yellow';">如果迭代次数过高，拟合的准确性往往会降低</span></u>，可以比较一下迭代<u><span style="color:'green';">1000</span></u>次和迭代<u><span style="color:'green';">500</span></u>次的R2值，发现迭代<u><span style="color:'green';">1000</span></u>次的拟合比迭代<u><span style="color:'green';">500</span></u>次的稍微差一点
